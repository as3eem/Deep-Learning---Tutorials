```
#training on our data
eng_text = open('/content/drive/MyDrive/NLP_assignment/assignment04/data/english.txt').read()
eng_text
```

```
#preprocessing the data
text = re.sub(r'\[[0-9]*\]',' ',eng_text)
text = re.sub(r'\s+',' ',text)
text = text.lower()
text = re.sub(r'\d',' ',text)
text = re.sub(r'\s+',' ',text)
text = re.sub(r'[\,-]','',text)
text = re.sub(r'.?[\'@#!~%*()/{}\-\*\+]','',text)
print(text)
```

```
#preparing the dataset
sentences = nltk.sent_tokenize(text)
sentences = [nltk.word_tokenize(sentence) for sentence in sentences]
#get sentences
for i in range(len(sentences)):
    sentences[i] = [word for word in sentences[i] if word not in stopwords.words('english')]
 
```

HINDI Preprocessing

```
#training on our data
hi_text = open('/content/drive/MyDrive/NLP_assignment/assignment04/data/hindi.txt').read()

#preprocessing the data
text = re.sub(r'\[[a-zA-Z][0-9]*\]',' ',hi_text)
text = re.sub(r'\s+',' ',text)
text = text.lower()
text = re.sub(r'\d',' ',text)
text = re.sub(r'\s+',' ',text)
text = re.sub(r'[\,-]','',text)
text = re.sub(r'.?[\'@#!~%*()/{}\-\*\+]','',text)
```

```
# indic library
!pip install indic-nlp-library
```

```
from indicnlp.tokenize import indic_tokenize  

#preparing the dataset using indic
hindi_sentences = re.split('\।', text) 
hindi_tokens = [indic_tokenize.trivial_tokenize(sentence) for sentence in hindi_sentences]
```

```
hindi_stopwords_file = open("/content/drive/MyDrive/NLP_assignment/assignment04/hindi-stopwords.txt")
hindi_stopwords = set()
for line in hindi_stopwords:
  line1 = line.strip("\n")
  hindi_stopwords.add(line1)
for i in range(len(hindi_tokens)):
    hindi_tokens[i] = [word.lower() for word in hindi_tokens[i] if word not in hindi_stopwords]
```

```
file1 = open("/content/drive/MyDrive/NLP_assignment/assignment04/glove-input-english.txt", "w")
file2 = open("/content/drive/MyDrive/NLP_assignment/assignment04/glove-input-hindi.txt", "w")

for sentence in sentences:
  for word in sentence:
    file1.write(word + " ")

for sentence in hindi_tokens:
  for word in sentence:
    file2.write(word + " ")
```

```
from glove import Corpus, Glove

#Creating a corpus object
corpus = Corpus() 
#Training the corpus to generate the co-occurence matrix which is used in GloVe
corpus.fit(sentences, window=10)
```

```
glove = Glove(no_components=50, learning_rate=0.05) # 50 is the embedding dimension

glove.fit(corpus.matrix, epochs=30, no_threads=4, verbose=True)
glove.add_dictionary(corpus.dictionary)

glove.save('/content/drive/MyDrive/NLP_assignment/assignment04/trained-models/50_glove_model_eng.model')

glove.word_vectors[glove.dictionary['computer']]
```

GLOVE Hindi

```
#Creating a corpus object
corpus = Corpus() 
#Training the corpus to generate the co-occurence matrix which is used in GloVe
corpus.fit(hindi_tokens, window=10)

glove = Glove(no_components=50, learning_rate=0.05) # 50 is the embedding dimension

glove.fit(corpus.matrix, epochs=30, no_threads=4, verbose=True)
glove.add_dictionary(corpus.dictionary)

glove.save('/content/drive/MyDrive/NLP_assignment/assignment04/trained-models/50_glove_model_hindi.model')
```
